 ### In this project, we were given a list of words with multiple possible classes. 
 ### No features have been given, so the task is to generate our own.
 ### Besides basic features such as length & pre/suf-fix, I wanted to experiment with OpenAI's text-davinci API.

```python
#Imports
import pandas as pd
import numpy as np
import matplotlib
import os
import re
import openai
openai.api_key = 'OPENAI_KEY'
```

```python
#Loading Data
data=pd.read_csv('Part_Of_Speech.csv')
data.head()
```

```python
#Using OpenAI to find mood of words
p="Classify the mood in these words:\n"
inputs=data['Word']
mood=[]
#Iterating through words
for i in range(0,inputs.shape[0],10):
    #Parsing words into blocks of 10
    words=""
    twords=""
    for j in range(10):
        if((i+j)<2617):
            pos=i+j
            #Formatting words into format: 1. word\n2. word\n3. word\n......10.word\n
            words+=(str(j+1)+". "+inputs.iloc[pos]+"\n")
            #Formatting words for display
            twords+=(str(pos)+"-"+inputs.iloc[pos]+" ")
    #Using Davinci to generate response for prompt+(Block of words)
    response = openai.Completion.create(
        model="text-davinci-003",prompt=p+words,
        temperature=0,max_tokens=240,top_p=1,
        frequency_penalty=0,presence_penalty=0)
    #Extracting only response
    temp=response["choices"][0]["text"]
    #Splitting each word into list of 10
    temp=temp.split("\n")
    #First entry in list is blank
    del temp[0]
    print(twords)
    #Condensing possible outputs to 4 options
    for k in range(len(temp)):
        if 'Neutral' in temp[k]:
            temp[k]='Neutral'
        elif 'Positive' in temp[k]:
            temp[k]='Positive'
        elif 'Negative' in temp[k]:
            temp[k]='Negative'
        else:
            temp[k]='Other'
    #Preventing DaVinci glitching from breaking formatting
    if(len(temp)!=10 and i!=2610):
        dif=10-len(temp)
        for n in range(dif):
            temp=temp+['Other']
    mood+=temp
    print(temp,len(temp),len(mood))
```

```python
data.insert(1,'Mood',mood, True)
```

```python
#Using OpenAI to find syllable count
p="Count how many syllables do these words have:\n"
inputs=data['Word']
syl=[]
#Iterating through words
for i in range(0,inputs.shape[0],10):
    #Parsing words into blocks of 10
    words=""
    twords=""
    for j in range(10):
        if((i+j)<2617):
            pos=i+j
            #Formatting words into format: 1. word\n2. word\n3. word\n......10.word\n
            words+=(str(j+1)+". "+inputs.iloc[pos]+"\n")
            #Formatting words for display
            twords+=(str(pos)+"-"+inputs.iloc[pos]+" ")
    #Using Davinci to generate response for prompt+(Block of words)
    response = openai.Completion.create(
        model="text-davinci-003",prompt=p+words,
        temperature=0,max_tokens=1000,top_p=1,
        frequency_penalty=0,presence_penalty=0)
    #Extracting only response
    temp=response["choices"][0]["text"]
    #Splitting each word into list of 10
    temp=temp.split("\n")
    #First entry in list is blank
    del temp[0]
    print(twords)
    #Cleaning results
    for k in range(len(temp)):
        temp[k]=temp[k][3:]
        num=re.findall(r'\d+',temp[k])
        if(len(num)>0):
            temp[k]=int(num[0])
        else:
            temp[k]=-1
    #Preventing DaVinci glitching from breaking formatting
    if(len(temp)!=10 and i!=2610):
        dif=10-len(temp)
        for n in range(dif):
            temp=temp+[-1]
    syl+=temp
    print(temp,len(temp),len(syl))
```

```python
data.insert(1,'Syl_Count',syl,True)
data.head()
```

```python
#Fixing missing value in Syl_Count by replacing:
# -1 with rounded length/4 as an estimate
for i in range(data.shape[0]):
    if(data['Syl_Count'].loc[i]==-1):
        data['Syl_Count'].loc[i]=round( len(data['Word'].loc[i])/4 )
```
